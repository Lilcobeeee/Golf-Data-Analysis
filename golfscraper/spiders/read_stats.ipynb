# Manipulating the Dataframes
In this section I will be using the file I created with the categories of interest to sort out the categories from the scraped data. This way I am only working with relevant data for my analysis. I will add comments in the code along the way to describe what I am doing. <br> First I am creating a Dataframe with the categories of interest. I will use this Dataframe to sort the original Dataframe

```python
names = stats_of_interest['Title']
names = pd.DataFrame(names)
```

Lowering and changing the datatypes of the two dataframes so I can compare the categories with ease.

```python
names['Title'] = names['Title'].apply(lambda x: x.lower())
```

```python
golf_html['stat_cat'] = golf_html['stat_cat'].astype('str')
```

```python
golf_html['stat_cat'] = golf_html['stat_cat'].apply(lambda y: y.lower())
```

Below I am using the Dataframe I created to compare that all 55 categories are stored within the original dataset. <br>
With For loop I am also creating a DataFrame that I will use to sort the original scraped data.

```python
stats = []
for titles in golf_html['stat_cat']:
    for name in names['Title']:
        if titles == name:
            stats.append(titles)
            print(titles)
        elif titles != name:
            pass
    
```

Below is probably the most important set of code for this notebook. This is what will create the large DataFrame to be used for Data Analysis. I will do my best to step through line by line within the code <br>

```python
i = 0                                                                           #Creating Iterator for the Categories
x = 0                                                                           #Creating Iterator for the Dataframes
for dfs in gf_final['stat_table']:                                              #Initially I create a For Loop that loops through each row in the Dataframe on the tables 
    #print(dfs)
    df_test = pd.read_html(dfs)                                                 #First I need to convert the table from HTML to a readable format
    #print(df_test)
    df_frame = pd.DataFrame(df_test[0])                                         #Because of the way it is converted the Dataframe is 3D. I am only interested in the first index. I use this to create the dataframe for that specific row
    
    for q in df_frame.columns[3:]:                                              #From there I use the newly created DataFrame to add the Categories and the Tab Name to each relevant column. I seperate with --- so I can remove the names if need be later
        Y = q + '---' + gf_final['stat_cat'][i] + '---' + gf_final['group'][i]  #Line where I am adding the naming convention, note the i iterator, this must be inline with the row so I am not mixing up data or anything else. 
        df_frame.rename(columns = {q : Y}, inplace = True)                      #Renameing the columns accordingly
    df_frame.drop(df_frame.columns[[0,1]],axis =1, inplace=True)                #Dropping the first two columns (rank information) because they aren't needed

    i = i+1                                                                     #Dont forget to increase my i
    if x == 0:                                                                  #This if statement is for building the large Dataframe with all the data. First I check to see what Index I'm on. If 0 then I know I'm just starting
        df_final = df_frame                                                     #From here we create our final Dataframe - df_final - this DataFrame will store all of the stats and player information
        x = x+1                                                                 #Increase my iterator 
        pass                                                                    #Maybe uneeded but we'll leave it
    elif x >= 0:                                                                #Now we check to see if x is greater than or equal to 0, if True then we know were on our second dataframe and we need to combine it with our df_final
        df_final = df_final.merge(df_frame, on ='PLAYER NAME')                  #Take the Dataframe and merge with df_final to create a larger Dataframe. Continue through the dataset and we have a very large table
        x = x+1                                                                 #Don't forget the iterator!
```
